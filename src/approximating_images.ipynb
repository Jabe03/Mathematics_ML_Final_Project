```python
data_dict = torch.load("MNIST_images_size_2048")
images = data_dict["images"]
labels = data_dict["labels"]
BATCH_SIZE = data_dict["BATCH_SIZE"]
```

```python



ratio = 1/20
CUTOFF = floor(BATCH_SIZE * ratio)

images_train = images[:CUTOFF]
images_test = images[CUTOFF:]
print(f"Images dataset={len(images_train)}, Images test={len(images_test)}")

```

```python
image_aproximater = utilities.get_feature_approximation_fn(images_train)
```
